Let's take the code we came up with from our pseudocode and try the first round of implementation 

```python
from numpy import random as rnd

flag = True

while flag:
    
    num1 = rnd.randint(1000)
    num2 = rnd.randint(1000)
    
    usr_in = int(input(f'Hey! Can you solve this? \n \n  {num1:>3}\n+ {num2:>3}\n----- \n'))
    
    if usr_in == num1+num2:
        
        usr_in = input("Congratulations! Press any key to try another, or enter 'Q' to quit: ")
        
        if usr_in == "Q" or usr_in == 'q':
            
            flag = False
            
    else:
        
        usr_in = input("Oooof! Not quite. Press any key to try again or 'Q' to quit: ")
        
        if usr_in == "Q" or usr_in == 'q':

            flag = False
        else:

            usr_in = int(input(f'Here it is again \n \n  {num1}\n+ {num2}\n----- \n'))
    
```

This is a really good first implementation! It at least has the most basic functionality for the most part. The main issue here is that it only allows a user to get a problem wrong once, and then regardless what they do on the second try it will just present a new problem. Instead, it should evaluate the answer and then proceed down one of two paths, **always**. The solution here might be to add some modularity to our code. For example, have a separate module (i.e. a function) for evaluating the answers and proceeding down one of two paths depending if they got it right or wrong.

### Let's see if we can get the pseudocode for this:

1. Present a math problem

    - collect user input

2. Determine if correct or not

    - if correct, congratulate and ask if they want to go again or quit
        - if go again, iterate through while loop
            - else, flag -> false and terminate
            
    - else
        - if go again, present same problem (don't iterate)
            - else, flag -> false and terminate
            
            
This looks like we should have a function which presents a math problem and collects user input. We also want to use recursion to re-call this function in the case where the use gets the question wrong, but wants to try the same problem again.

```python
from numpy import random as rnd

def get_answer(num1, num2):
    
    usr_in = int(input(f'Hey! Can you solve this? \n \n  {num1:>3}\n+ {num2:>3}\n----- \n'))
    
    if usr_in == num1+num2:
        
        usr_in = input("Congratulations! Press any key to try another, or enter 'Q' to quit: ")
        
        if usr_in == "Q" or usr_in == 'q':
            
            return False
        
        else:
            return True
            
    else:
        
        usr_in = input("Oooof! Not quite. Press any key to try again or 'Q' to quit: ")
        
        if usr_in == "Q" or usr_in == 'q':
            
            return False
        
        else:
            
            return get_answer(num1, num2)

flag = True
        
while flag:
    
    # generate the random numbers
    num1 = rnd.randint(1000)
    num2 = rnd.randint(1000)
    
    flag = get_answer(num1, num2)
```

This fixed our previous issue, but now we see another: the program crashes if the user enters a non-integer value for the answer to the problem. We need to add in some exception handling.

```python
from numpy import random as rnd

def get_answer(num1, num2):
    
    usr_in = input(f'Hey! Can you solve this? \n \n  {num1:>3}\n+ {num2:>3}\n----- \n')
    
    try:
        usr_in = int(usr_in)
        
    
        if usr_in == num1+num2:

            usr_in = input("Congratulations! Press any key to try another, or enter 'Q' to quit: ")

            if usr_in == "Q" or usr_in == 'q':

                return False

            else:
                return True

        else:

            usr_in = input("Oooof! Not quite. Press any key to try again or 'Q' to quit: ")

            if usr_in == "Q" or usr_in == 'q':

                return False

            else:

                return get_answer(num1, num2)
    except:

        print("\n\nPlease enter integer values only!")

        return get_answer(num1, num2)
        

flag = True
        
while flag:
    
    # generate the random numbers
    num1 = rnd.randint(1000)
    num2 = rnd.randint(1000)
    
    flag = get_answer(num1, num2)
```

^^^Now, this works. We found the issue by using the Python Tutor. We saw that when inside the except branch we were setting the flag variable equal to the function call, but we were not returning a value! Therefore, the return value to the first function was type None, and the flag variable was being set to the return value of the first function call, therefore, whenever we were on the except branch, flag was being set to None and causing us to the exit the while loop.
